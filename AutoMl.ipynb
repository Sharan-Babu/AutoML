Hi. Today we are going to se how to use AutoViML which is a library used for Automating the machine learning process to an 
extent. It is capable of providing data visualizations after automatically cleaning and filtering the data. It tests different models and finally uses the best model to fit on the train data. It returns plots for ROC curves, Precision-Recall curve and other useful visualization plots. Finally, it returns the classification report of the model. 

This kind of automated machine learning library can be used to quickly get a rough idea of the best model for a given data set and also get an idea of the benchmark accuracy score. Inspite of all the amazing things this library can quickly do, it still can't do useful stuff like feature engineering.

Split your data set into train and test sets.<br>
I will be using train_df and test_df as my datasets. I am using the Titanic dataset.<br>
Check: https://github.com/AutoViML/Auto_ViML for documentation.

```python
# Import the necessary libraries
import numpy as np
import pandas as pd

from autoviml.Auto_ViML import Auto_ViML
```

```python
# Split data into train set and test set
train = pd.read_csv(r"C:\Users\Sharan Babu\Desktop\titanic_train.csv")
train_df = train[:623]
test_df = train[623:]
```

```python
# taregt column or column to be predicted is 'Survived'
target='Survived'
```

```python
# Set the parameters. Trained and Tested are names of the dataframes on which we can use to see individual model
# performances
model, features, trained, tested = Auto_ViML(train_df, target, test_df, sample_submission='',
                                    scoring_parameter='',
                                    hyper_param='RS',feature_reduction=True,
                                     Boosting_Flag=True,Binning_Flag=False,
                                    Add_Poly=0, Stacking_Flag=False, 
                                    Imbalanced_Flag=True, 
                                    verbose=1)

# Many different kinds of plots are returned showing which features have more importnace in predicting the outcome.
# Multiple Visualizations
# Data fit using multiple models. 
# We can use the best model fit for our purposes.
```
